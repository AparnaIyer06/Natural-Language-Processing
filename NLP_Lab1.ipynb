```python
print(text.similar('Moby'))
```

```python
counts=nltk.FreqDist(text) #Dictionary
vocab = len(counts.keys()) #How many unique words
words = sum(counts.values()) #How many words
lexdiv = float(words)/float(vocab) #How many unique words - Diversity but less understandability
```

```python
#Less Lexdiv (Lexical Diversity): Children (Readable or understandable by younger audiences)
#Medium Lexical Diversity:
#Used by Editors, Writers
#Low Lexical Diversity: Simple
print(lexdiv)
```

```python
counts.most_common(10)
```

```python
tags = []
for sent in nltk.sent_tokenize(txt1):
  tags.append(nltk.pos_tag(nltk.word_tokenize(sent))) #POS Tagging using NLTK Library- Identified using Penn Treebank
```

```python
print(tags[8]) #Each sentence in the text
```
